```python
movies = pd.read_csv("movies.dat", sep='\t', header=0, encoding='latin-1') # set the encoding: required for this file
movie_genres = pd.read_csv("movie_genres.dat", sep='\t', header=0)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Initialize the OneHotEncoder
encoder = OneHotEncoder(sparse=False)

# Fit and transform the genre data
encoded_genres = encoder.fit_transform(movie_genres[['genre']])

# Convert the encoded genres to a DataFrame with appropriate column names
list_categories = encoder.categories_[0].tolist()
encoded_genres_df = pd.DataFrame(encoded_genres, columns=list_categories)

# Combine the movie IDs with the one-hot encoded genres
genres_one_hot = pd.concat([movie_genres[['movieID']], encoded_genres_df], axis=1)

# Aggregate by movie ID and convert one-hot vectors to strings
aggregated_genres = genres_one_hot.groupby('movieID').sum().reset_index()
```

```python
aggregated_genres.head()
```

```python
vector1 = aggregated_genres.loc[0, list_categories].values
vector2 = aggregated_genres.loc[1, list_categories].values
```

```python
import scipy.spatial.distance as sp_dist

print(sp_dist.cosine(vector1, vector2))
print(sp_dist.euclidean(vector1, vector2))
print(sp_dist.jaccard(vector1, vector2))
```
